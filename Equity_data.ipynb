```python
import pandas as pd
import yfinance as yf
```

```python
# Define tickers
tickers = {
    "DAX": "^GDAXI", ## German
    "FTSE_MIB": "FTSEMIB.MI", ## Italuan
    "CAC_40": "^FCHI", ## France
    "IBEX_35": "^IBEX", ## Spain
    "Eurofirst 50": "^STOXX50E"  # Pan-European
}

data = []
for name, ticker in tickers.items():
    df = yf.download(ticker, start="2009-01-01", end="2025-09-30")
    series = df['Close']
    series.name = name  # safer than .rename()
    data.append(series)

# Combine all series into a single DataFrame aligned by date
all_indices = pd.concat(data, axis=1)

print(all_indices.head())
```

```python
all_indices2 = pd.DataFrame(all_indices)
all_indices2
```
